## Analyzing the Collisions in New York City

The data is publically available on Google BigQuery. I ran the queries and saved the results on GCS. I can directly import it from there now. 

```python
from google.cloud import storage
from io import BytesIO
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn
import warnings
warnings.filterwarnings('ignore')

bucket_name = "graphx-usage"
```

```python
!gsutil cp gs://graphx-usage/all_accidents.csv data/accidents_full.csv
```

```python
!gsutil cp gs://graphx-usage/accident.csv data/accidents.csv
```

```python
accident = pd.read_csv('data/accidents_full.csv')
```

```python
accidents = accident[['borough', 'timestamp', 'latitude', 'longitude']]
```

```python
accidents = accidents[(accidents['latitude'] >= 40) & (accidents['longitude'] < 0)].reset_index().iloc[:, 1:]
```

```python
def convert_ddy(timestamp):
    sp = timestamp.split(' ')
    date = sp[0].split('-')
    year, month, day = int(date[0]), int(date[1]), int(date[2])
    return year, month, day
```

```python
y = []
m = []
dy = []

for i in range(len(accidents)):
    year, month, day = convert_ddy(accident['timestamp'][i])
    y.append(year)
    m.append(month)
    dy.append(day)
```

```python
len(y), len(m), len(dy), accidents.shape
```

```python
accidents['year'] = y
accidents['month'] = m
accidents['day'] = dy
```

```python
accidents.head()
```

```python
accidents['borough'].unique()
```

```python
bdict = {'BROOKLYN': 1, 'MANHATTAN': 2, 'QUEENS': 3, 'STATEN ISLAND': 4, 'BRONX': 5}
```

```python
barr = []
for i in range(len(accidents)):
    if(accidents['borough'][i] in bdict.keys()):
        barr.append(bdict[accidents['borough'][i]])
    else:
        barr.append(0)
```

```python
accidents['borough_encode'] = barr
```

```python
accidents = accidents[['year', 'month', 'day', 'borough_encode', 'latitude', 'longitude']]
```

```python
accidents.to_csv('data/accidents.csv')
```

## The Data was processed on Cloud Dataproc and Clustered using KMeans Clustering Algorithm. Below are the visualizations for the Clusters.

```python
!gsutil cp gs://graphx-usage/clusters/part-00000-843aecef-3e47-47ab-b778-7c307eb282de-c000.csv data/cluster_data.csv
```

```python
#getting actual data from BQ
from google.cloud import bigquery
client = bigquery.Client()

sql = """
SELECT *
from `summerai.nyc_mv_collisions_processed.collision_data`
where int64_field_0 is not null
and year is not null
and month is not null
and day is not null
and borough_encode is not null
and latitude is not null
and longitude is not null
"""
df = client.query(sql).to_dataframe()
```

```python
df.shape
```

```python
df.columns = [u'label', u'year', u'month', u'day', u'borough_encode',
       u'latitude', u'longitude']
```

```python
dfclust = pd.read_csv('data/cluster_data.csv')
```

```python
fdf = pd.merge(df, dfclust, on='label')
```

```python
fdf.head()
```

## Visualization 

```python
import geopandas as gpd
from shapely.geometry import Point, Polygon
import descartes
```

```python
street_map = gpd.read_file('ShapeFiles/nyu_2451_34505.shp')
```

```python
geometry = [Point(xy) for xy in zip(fdf['longitude'], fdf['latitude'])]
crs = {'init':'epsg:4326'}
geodf = gpd.GeoDataFrame(fdf, crs=crs, geometry=geometry)
```

```python
geodf['prediction'].unique()
```

```python
geodf2 = geodf.sample(int(185259/15))
```

```python
#Unique Colors
NUM_COLORS = 10
cm =['black', 'blue', 'brown', 'red', 'grey', 'y', 'orange', 'cadetblue',\
 'cyan', 'pink', 'yellow', 'linen', 'm', 'beige']
dd = dict()
for i in range(NUM_COLORS):
    dd[i] = cm[i]
    
marker = ['o', 'o', 'o']
    

fig, ax = plt.subplots(figsize=(15, 15))
street_map.plot(ax=ax, alpha=0.4, color='beige')
for i in range(NUM_COLORS):
    try:
        geodf[geodf['prediction'] == i]['geometry'].plot(ax=ax, color=dd[i], marker='o', label=i, alpha=0.5)
    except:
        pass

minx, miny, maxx, maxy = -74.20, 40.55, -73.70, 40.90
ax.set_xlim(minx, maxx)
ax.set_ylim(miny, maxy)

plt.title('Clustered Points')
plt.legend()
plt.show()
```
